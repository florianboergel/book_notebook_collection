# Accessing EMODNET sea level data from tide gauges

The sea level data can be downloaded via the Emodnet ERDDAP server like this:

```python
import urllib.request
import json 
import requests
import os

os.chdir("/silos/notebook_collection/accessing_data/sealevel_emodnet/")
  
###### configuration variables ######
savepath="/data/sealevel/emodnet_hourly/"     ### folder where you want to save the files
fileType=".nc"            ### see https://erddap.emodnet-physics.eu/erddap/tabledap/documentation.html#fileType 
min_latitude="50.0"    ### min latitude used to retrieve platforms
max_latitude="67.0"     ### max latitude used to retrieve platforms
min_longitude="-5.0"   ### min longitude used to retrieve platforms
max_longitude="31.0"   ### max longitude used to retrieve platforms


###### main ######
uri_distinct_platforms="https://erddap.emodnet-physics.eu/erddap/tabledap/EP_ERD_INT_SLEV_AL_TS_NRT.json?EP_PLATFORM_ID%2CEP_PLATFORM_CODE&latitude%3E=" + min_latitude + "&latitude%3C=" + max_latitude + "&longitude%3E="+ min_longitude + "&longitude%3C=" + max_longitude + "&distinct()"
with urllib.request.urlopen(uri_distinct_platforms) as url:
    data = json.loads(url.read().decode())   
for platform in data["table"]["rows"]:
    uri="https://erddap.emodnet-physics.eu/erddap/tabledap/EP_ERD_INT_SLEV_AL_TS_NRT" + fileType + "?&EP_PLATFORM_ID=%22" + platform[0] + "%22"
    filename = os.path.join(savepath,"EP_ERD_INT_SLEV_AL_TS_NRT_" + platform[1] + fileType)
    try:
        print("Donwloading data for platform ",platform[1]," ... ", end="", flush=True)
        urllib.request.urlretrieve(uri, filename)
        print("done!")
    except Exception as e:    
        print("Error donwloading -> ",e)
```
